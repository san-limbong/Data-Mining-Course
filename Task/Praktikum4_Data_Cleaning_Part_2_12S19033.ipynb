### Latihan 2.4 | Membersihkan Kolom dengan Kombinasi metode str dengan NumPy

Kode program berikut berguna untuk membersihkan field Place of Publication, dengan kombinasi metode str di Panda dengan fungsi np.where di NumPy yang serupa dengan bentuk vektor dari makro IF() di Excell,

```python
#np.where(condition, then, else)
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                              pub.str.replace('-', ' ')))
df['Place of Publication'].head()
```

### Latihan 2.5 | Membersihkan Seluruh Dataset dengan Fungsi applymap

Metode.applymap() yang menerapkan fungsi built-in yaitu fungsi map() berguna untuk membersihkan seluruh dataset

```python
university_towns = []
with open("C:/Users/ASUS/a semester 7/DAMI/New folder/data/data/university_towns.txt") as file:
    for line in file: 
        if '[edit]' in line:
            #remember this 'state' until the next is found
            state = line
        else:
            #otherwise, we have a city; keep 'state' as last-seen
            university_towns.append((state, line))

university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns,
                       columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(get_citystate)
```

```python
towns_df.head()
```
